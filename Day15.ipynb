## Types of errors in python

- Syntax Error:

and error that occurs in syntaxing or writing code mistake

example: print 'Hello world' instead of print("Hello world")


 - Name error

 This is the error when that occurs when trying to execut undefined variable
 ```sh
 asabeneh@Asabeneh:~$ python
Python 3.9.6 (default, Jun 28 2021, 15:26:21)
[Clang 11.0.0 (clang-1100.0.33.8)] on darwin

>>> print(age)

  File "<stdin>", line 1, in <module>
NameError: name 'age' is not defined

>>> print(age)
```
