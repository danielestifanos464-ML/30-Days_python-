```python
raw_scores = [" 78", "56", " ninety", "88 ", "45marks", " 67"]

clean_scores = []
invalid_ = 0

for score in raw_scores:
    try:
        num = int(score.strip())
        clean_scores.append(num)
    except:
        invalid_ += 1


average = round(sum(clean_scores) / len(clean_scores), 2)

highest = max(clean_scores)


print("Clean Scores of integers:-", clean_scores)
print("Invalid Entries:-", invalid_)
print("Average Score:-", average)
print("Highest Score:-", highest)
    
```

```python
try:
    score = float(input("Enter exam score (0–100): "))
    attendance = float(input("Enter attendance (0–100): "))
    #range
    if not (0 <= score <= 100) or not (0 <= attendance <= 100):
        print("Error: Inputs must be between 0 and 100.")
    else:
        # Additional rule
        if score == 100 and attendance == 100:
            print("Congratulations! Perfect performance bonus!")

        # Scholarship
        if score >= 75 and attendance >= 80:
            print("Scholarship: Full")

        elif score >= 60 and attendance >= 70:
            print("Scholarship: Partial")

        else:
            print("Scholarship: None")

        # Attendance
        if attendance < 50:
            print("Warning: Low attendance!")

except ValueError:
    print("Invalid input! Please enter numeric values only.")
```

This task applies conditional logic to determine scholarship eligibility based on validated exam score and attendance inputs.

- Question 3

This program calculates total, average, counts above-average values, and identifies the lowest value before updating and recalculating the results.

```python
# Create list of 8 daily visits
visits = [120, 150, 90, 200, 175, 80, 160, 140]
#total_sum
total = 0
for v in visits:
    total += v
# average
average = total / len(visits)
#days above average
above_avg_count = 0
for v in visits:
    if v > average:
        above_avg_count += 1
#Find lowest value manually
lowest = visits[0]
for v in visits:
    if v < lowest:
        lowest = v
# Print results
print("Visits:", visits)
print("Total Visits:", total)
print("Average Visits:", round(average, 2))
print("Days Above Average:", above_avg_count)
print("Lowest Visits:", lowest)
#Change one value midway
print("\n***Changed_value from 90 ti 130\n")
visits[2] = 130   # changed from 90 to 130
#Re-run calculations
total = 0
for v in visits:
    total += v
average = total / len(visits)
above_avg_count = 0
for v in visits:
    if v > average:
        above_avg_count += 1
lowest = visits[0]
for v in visits:
    if v < lowest:
        lowest = v
# Print updated results
print("Updated Visits:", visits)
print("New Total:", total)
print("New Average:", round(average, 2))
print("New Days Above Average:", above_avg_count)
print("New Lowest:", lowest)
```

- Question 4

this program converts the intial sentence into a list and applies transformations like counting, replacing, and slicing.

```python
#sentence
sentence = "Data science combines statistics programming mathematics and machine learning to extract insights from data and helps to make calculated decisions"

#Convert to uppercase
upper_sentence = sentence.upper()

#Convert sentence into list using method
words = sentence.split()

#Count number of words
word_count = len(words)

#Count words longer than 6 characters
long_word_count = 0
for w in words:
    if len(w) > 6:
        long_word_count += 1

#Replace one word
new_sentence = sentence.replace("statistics", "analytics")

#Slice first 10 characters
first_10 = sentence[:10]

#Additional transformation (reverse the sentence)
reversed_sentence = sentence[::-1]

# Print results
print("----Original Sentence:", sentence,'\n')
print("----Uppercase:", upper_sentence,'\n')
print("----Word Count:", word_count,'\n')
print("----Words Longer Than 6 Characters:", long_word_count,'\n')
print("----After Replacement:", new_sentence,'\n')
print("----First 10 Characters:", first_10,'\n')
print("----Reversed Sentence:", reversed_sentence,'\n')
```

- Question 5

This program calculates the average temperature and counts the number of days above 30degree celsius using loops and conditions. Based on the results, it classifies the period and applies an additional heat alert rule for interpretation.

```python
# Temperature
Estif_temperatures = [29, 31, 27, 33, 30, 26, 34, 28, 32, 25]

#average
Estif_total = 0
for temp in Estif_temperatures:
    Estif_total += temp

Estif_average = Estif_total / len(Estif_temperatures)

#above 30
Estif_above30 = 0
for temp in Estif_temperatures:
    if temp > 30:
        Estif_above30 += 1

######
if Estif_average > 28:
    print("Hot_Period")
else:
    print("Normal_Period")
######
if Estif_above30 >= 4:
    print("High Heat Alert: Several extreme temperature days detected.")

#results
print("Temperature Readings:", Estif_temperatures)
print("Average Temperature:", round(Estif_average, 2))
print("Days Above 30oc:", Estif_above30)
```
