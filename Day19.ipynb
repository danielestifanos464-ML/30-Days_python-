````python
import requests
import re
url = ("https://raw.githubusercontent.com/Asabeneh/30-Days-Of-Python/refs/heads/master/data/melina_trump_speech.txt")
response = requests.get(url)
text = response.text

#coping it to my folder
with open("C:\\Users\\hp eiltebook 1040G8\\OneDrive\\Desktop\\30 Days Python\\melina_trump_speech.txt.txt", "w", encoding="utf-8") as f:
    f.write(text)
#reading it again all
with open("C:\\Users\\hp eiltebook 1040G8\\OneDrive\\Desktop\\30 Days Python\\melina_trump_speech.txt.txt","r") as f:
    file = f.read()
#defining the function
def find_most_common_words(txt,num):
    num = int(num)
    text = re.sub(r'[^a-zA-Z\s]', ' ', txt.lower()).split()
    dict_words ={}
    for i in text:
        if i in dict_words:
            dict_words[i] +=1
        else:
            dict_words[i] = 1
    listed_words = sorted(dict_words.items(), key=lambda x:x[1],reverse=True)
    return listed_words[:num]

#out_put
print(find_most_common_words(file,10))
print(f"The number of words are {len(file)}")
    
````

```sh
Write a program that:
•	Accepts 5 numbers from the user
•	Stores them in a list
•	Displays:
o	The maximum
o	The minimum
o	The average
```


````python
def formula(a,b,c,d):
    list_=[a,b,c,d]
    sorted(list_)
    Average = sum(list_)/len(list_)
    Maximum = max(list_)
    Minmum = min(list_)
    return sorted(list_),Average,Maximum,Minmum
````

````python
formula(1,4,3,2)
````
