1. Three of them are higher order in built functions they take a function and interable and do the task of the function in the iterable
    - map is takes and function and transforms all the elements in the iterable
    - filter takes the function as argument and filters the elements based on the command
    - reduce take the function and gives as one single value

2.   - higher order functions are functions that takes or returns functionsas parameters
     - closure is a function that is built inside a functions and remembers the varibale in the outer function even after it finshes
     - decorators is a function that modify, or adds to a function inside it it has the command wrapper inside a innner function(which is like closure)

````python
#Create a function returning a dictionary, where keys stand for starting letters of countries and values are the number of country names starting with that letter.


#****just for practicing decorators****
def dectry(function):
    
    def waraper(x):
        unique_list = []
        fun = function(x)
        for i in fun :
            if i not in unique_list:
                unique_list.append(i)
        return unique_list
    return waraper
    
@dectry
def letter_and_count(x):
    first_letter = []
    for i in range(len(x)):
        a = x[i]
        first_letter.append(a[0])
    return first_letter


````

````python
#Create a function returning a dictionary, where keys stand for starting letters of countries and values are the number of country names starting with that letter.

def counter(x_list):
    count_dict = {}

    for i in x_list:
        country= i[0]
        if country in count_dict:
            count_dict[country] += 1
        else:
            count_dict[country] = 1

    return count_dict
````
