Write a code which gives grade to students according to theirs scores:
```sh
90-100, A
80-89, B
70-79, C
60-69, D
0-59, F
```

`````python
grade = round(int(input('put students grade for labeling')),0)
if grade <=100 and grade>=90:
    print(f'Grade is {grade} which is labeled as A')
elif grade <=89 and grade>=80:
    print(f'Grade is {grade} which is labeled as B')
elif grade <=79 and grade>=70:
    print(f'Grade is {grade} which is labeled as C')
elif grade <=69 and grade>=60:
    print(f'Grade is {grade} which is labeled as D')
elif grade <=59 and grade>=0:
    print(f'Grade is {grade} which is labeled as F')
else:
    ('Grading might have some problem check with the registrar office')
`````

Get the month from user input then check if the season is Autumn, Winter, Spring or Summer. 
```sh
If the user input is: 
September, October or November, the season is Autumn. 
December, January or February, the season is Winter. 
March, April or May, the season is Spring 
June, July or August, the season is Summer
```


`````python
Month = (str(input('write What month we are in '))).capitalize()

if Month in ['September', 'October','November']:
    print('The season is Autumn')
elif Month in ['December', 'January','February']:
    print('The season is Winter')
elif Month in ['March', 'April','May']:
    print('The season is Spring')
elif Month in ['June', 'July','August']:
    print('The season is Summer')
else:
    raise KeyError('The month you typedis not found check for spelling error')
`````
