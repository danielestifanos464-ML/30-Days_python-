# OS Module

````python
import os
````

Using python os module it is possible to automatically perform many operating system tasks. The OS module in Python provides functions for creating, changing current working directory, and removing a directory (folder), fetching its contents, changing and identifying the current directory.

```sh
Core things os is used for (90% of usage)
📁 Directories

create → os.mkdir(), os.makedirs()

delete → os.rmdir(), shutil.rmtree()

move into → os.chdir()

list → os.listdir()

📄 Files

check existence → os.path.exists()

rename/move → os.rename()

delete → os.remove()

🧭 Paths

join paths → os.path.join()

absolute path → os.path.abspath()

file/dir name → os.path.basename(), os.path.dirname()

⚙️ System info (less common)

current directory → os.getcwd()

environment variables → os.environ

run command → os.system() (rarely used now)

```

# Module sys
- The sys module provides functions and variables used to manipulate different parts of the Python runtime environment. Function sys.argv returns a list of command line arguments passed to a Python script. The item at index 0 in this list is always the name of the script, at index 1 is the argument passed from the command line.

````python
import sys
````

```sh

```sh
sys operations and their uses

sys.argv
👉 Gets the words typed when the program is started
👉 Used for command-line programs

sys.exit()
👉 Stops the program immediately
👉 Used when something goes wrong or when the program should end

sys.version
👉 Shows the Python version being used
👉 Used for debugging and compatibility checks

sys.path
👉 Shows where Python looks for modules
👉 Used when imports fail or for custom module locations

sys.stdin
👉 Where input comes from
👉 Used for advanced input handling (pipelines, files)

sys.stdout
👉 Where normal output goes
👉 Used to redirect output to files or other programs

sys.stderr
👉 Where error messages go
👉 Used for logging errors separately from normal output

sys.getsizeof()
👉 Tells how much memory an object uses
👉 Used for performance and memory checks

```
