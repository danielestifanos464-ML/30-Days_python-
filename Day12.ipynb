```python
import random
import string
def random_user_id():
    user_id = ''
    strings_num = string.digits
    string_char = string.ascii_lowercase
    comp_user = strings_num + string_char

    for i in range(6):
       user_id += random.choice(comp_user)
        
    return user_id

 
```

```python
#Call your function shuffle_list, it takes a list as a parameter and it returns a shuffled list
def shuffle_list(*shuffle):
    list_ =[]
    for i in range(len(shuffle)-1,-1,-1):
        list_.append(shuffle[i])
    return list_
```
