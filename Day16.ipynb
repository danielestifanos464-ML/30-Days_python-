```python
#Get the current day, month, year, hour, minute and timestamp from datetime module
from datetime import datetime,date,time,timedelta
today = datetime.now() 
print(today)
#Format the current date using this format: "%m/%d/%Y, %H:%M:%S")
today_formated = datetime.strftime(today,"%m/%d/%Y, %H:%M:%S")
print(today_formated)

#Today is 5 December, 2019. Change this time string to time.
today_given ="5 December, 2019"
today_given  = datetime.strptime(today_given,'%d %B, %Y')
print(today_given )

#Calculate the time difference between now and new year.
today = datetime.now() 
new_year = datetime(day=1,month=1,year=2027)
time_diff = new_year-today
print(time_diff)
#Calculate the time difference between 1 January 1970 and now.
given_date = datetime(day=1,month=1,year=1970)
time_diff_with_given = given_date-today
print(time_diff_with_given)
```
